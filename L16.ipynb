```python
fig, ax = plt.subplots()  
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]); 
```

```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.set_title("this is my plot")
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);  # Plot some data on the axes.
```

```python
fig, ax = plt.subplots()  
ax.set_title("this is my plot")
ax.scatter([1, 2, 3, 4], [1, 4, 2, 3]);  
```

```python
fig, ax = plt.subplots() 
ax.plot([43, 5, 62, 12, 23, 3], [12, 23, 3, 43, 5, 62]);
ax.plot([12, 23, 3, 43, 5, 62], [12, 53, 22, 42, 12, 33])
```

```python
fig, ax = plt.subplots() 
ax.plot([43, 5, 62, 12, 23, 3], [12, 23, 3, 43, 5, 62], label = "first random line");
ax.plot([12, 23, 3, 43, 5, 62], [12, 53, 22, 42, 12, 33], label = "second random line")
ax.legend()
```
